```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data 
y = iris.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2, svd_solver='full')
X_pca = pca.fit_transform(X_scaled)

print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=100)
plt.xlabel('Primer Componente Principal')
plt.ylabel('Segundo Componente Principal')
plt.title('PCA en el Dataset Iris')
plt.colorbar(label='Clase')

plt.show()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
plt.figure(figsize=(8,6))
plt.bar(range(1, 4), explained_variance_ratio, alpha=0.7, align='center', label='Varianza explicada por componente')
plt.xlabel('Componentes Principales')
plt.ylabel('Ratio de Varianza Explicada')
plt.title('Ratio de Varianza Explicada por cada Componente Principal')
plt.xticks(range(1, 4))
plt.legend()
plt.show()
```
